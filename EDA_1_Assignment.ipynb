#Detecting and treating outliers

### Detecting outliers with the help of boxplots

```python
plt.figure(figsize=(10,5))
df1.boxplot()
```

### Treatment of outliers using winsorize

```python

from scipy.stats.mstats import winsorize
WinsorizedArrayLB = winsorize(df1.LB.values,(0.05,0.05))
plt.boxplot(WinsorizedArrayLB);
```

```python
WinsorizedArrayASTV = winsorize(df1.ASTV.values,(0.05,0.05))
plt.boxplot(WinsorizedArrayASTV);
```

```python
WinsorizedArrayMSTV = winsorize(df1.MSTV.values,(0.05,0.05))
plt.boxplot(WinsorizedArrayMSTV);
```

```python
WinsorizedArrayALTV = winsorize(df1.ALTV.values,(0.05,0.05))
plt.boxplot(WinsorizedArrayALTV);
```

```python
sns.histplot(np.sqrt(df1.ALTV))
```

```python
sns.boxplot(np.sqrt(df1.ALTV));
```

```python
# for ALTV category, using winsorization still gave us outliers.
# But by using the sqrt method, very few outliers are observed.
```

```python
WinsorizedArrayMLTV = winsorize(df1.MLTV.values,(0.05,0.05))
plt.boxplot(WinsorizedArrayMLTV);
```

```python
WinsorizedArrayWidth = winsorize(df1.Width.values,(0.05,0.05))
plt.boxplot(WinsorizedArrayWidth);
```

# Statistical Summary

```python
'''
2.	Statistical Summary:
●	Provide a statistical summary for each variable in the dataset, including measures of central tendency (mean, median) and dispersion (standard deviation, interquartile range).
●	Highlight any interesting findings from this summary.

'''
```

```python
df1.describe()
```

```python
# Measures of central tendency were found with the describe function in which mean, std ,and meadian(50%) were obseerved.
# The describe function doesn't give the IQR so that was calculated by creating a function. The difference between the 25 and 75 percentile is IQR
```

```python
IQR={}
for i in df1.columns:
  IQR[i] = df1[i].quantile(0.75) - df1[i].quantile(0.25)
IQR_df = pd.DataFrame(IQR,index=['IQR'])

pd.concat([df1.describe(),IQR_df])
```

```python
'''
 Findings:
 1. Columns like LB,UC,DS have mean and median almost same indicating that the data in those categories are almost normally distributed.
 2. IQR shows where 50% of the data lies.
 3. Columns like LB, ASTV, ALTV, MLTV, Width,MSTV and Tendency show moderate to high IQR values, indicating higher variability in these measurements.
 4. Columns like AC,FM,UC,DL show low IQR values, indicating low variability
 5. Columns like DS,DP,NSP show 0 as IQR value, indicating no variability
'''
```

# Data Visualization

```python
'''
3.	Data Visualization:
●	Create histograms or boxplots to visualize the distributions of various numerical variables.
●	Use bar charts or pie charts to display the frequency of categories for categorical variables.
●	Generate scatter plots or correlation heatmaps to explore relationships between pairs of variables.
●	Employ advanced visualization techniques like pair plots, or violin plots for deeper insights.
'''
```

```python
plt.figure(figsize = (8,4))
sns.histplot(WinsorizedArrayLB, bins = 20,kde = True)
plt.title('Histogram of Baseline Fetal Heart Rate (FHR)')
plt.xlabel('Baseline Fetal Heart Rate (FHR)')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize = (8,4))
sns.histplot(WinsorizedArrayASTV, bins = 20,kde = True)
plt.title('Histogram of Percentage of Time with Abnormal Short Term Variability in the FHR')
plt.xlabel('Percentage of Time with Abnormal Short Term Variability (ASTV) in the FHR.')
plt.ylabel('Frequency')
plt.show()
```

```python
# correlation heat map
```

```python
vars_to_plot = ['LB', 'AC', 'FM', 'UC', 'DL', 'DS', 'DP', 'ASTV', 'MSTV', 'ALTV', 'MLTV']


# Calculating the correlation matrix
correlation_matrix = df1[vars_to_plot].corr()

# Creating the correlation heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```

```python
# violin plot
```

```python
plt.figure(figsize=(12, 8))
for i, var in enumerate(vars_to_plot):
    plt.subplot(4, 3, i + 1)
    sns.violinplot(x=df1[var], palette='Set1')
    plt.title(var)
plt.tight_layout()
plt.show()
```
