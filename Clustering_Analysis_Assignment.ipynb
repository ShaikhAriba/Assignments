## Implementing Clustering Algorithms

### Agglomerative/Hierarchical

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
hc1 = AgglomerativeClustering(n_clusters = 4, linkage = 'ward')
```

```python
hc1.fit(scaled_df1)
```

```python
hc1_predict = hc1.fit_predict(scaled_df1)
```

```python
hc1_predict
```

```python
df['cluster_no'] = pd.DataFrame(hc1_predict,columns=['cluster_no'])
```

```python
hc2 = AgglomerativeClustering(n_clusters = 4, linkage = 'ward')
hc2.fit(scaled_df2)
hc2_predict = hc2.fit_predict(scaled_df2)
hc2_predict
df1 = pd.read_excel('/content/sample_data/EastWestAirlines.xlsx', sheet_name='data')
df1['cluster_no'] = pd.DataFrame(hc2_predict,columns=['cluster_no'])
```
