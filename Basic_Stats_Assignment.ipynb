# Standardization

Standardization is a technique used to convert big values to smaller ones which helps to reduce outliers and tries to normalize the distribution.
Standardization is done by using the formula: z = x - mu/sigma

```python
df.columns
```

```python
df1 = df.drop(['Date','Day','SKU','City','BU','Brand','Model'],axis = 1)
```

```python
df1.head()
```

```python
df1.shape
```

```python
from sklearn.preprocessing import StandardScaler   #Standardization
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(df1)
```

```python
df1_1 = scaler.transform(df1)
```

```python
df1_1 = pd.DataFrame(df1_1)
```

```python
df1_1.columns = ['Volume','AvgPrice','TotalSalesValue','DiscountRate%','DiscountAmount','NetSalesValue']
```

```python
df1_1.head()
```

```python
df1_1.shape
```

```python
sns.pairplot(df1_1)  #pairplot after standardizing
```

```python
sns.pairplot(df1)   #pairplot before Standardizing
```

With the help of pairplot, we compared the distibutions before standardization and after standardizing the data. We did not find any significant difference between both the plots.

# Conversion of Categorical Columns into Dummy Variables

We need to convert categorical data to dummy variables, because the model expects data only in numbers. But with the presence of categorical columns in our data, the model might not be able to understand it. To tackle this problem, we can convert our categorical data to numerical ones with the help of One-hot Encoding.

```python
df.head()
```

```python
# dataset with only categorical col
```

```python
df_cat = df.drop(['Date','Volume','Avg Price','Total Sales Value','Discount Rate (%)','Discount Amount','Net Sales Value'],axis = 1)
```

```python
df_cat.head()
```

```python
# one hot encoding on categorical columns
```

```python
pd.get_dummies(data = df_cat,dtype = int)
```

```python
# one-hot enc on a single categorical col
```

```python
pd.get_dummies(df_cat,columns=['SKU'], dtype = int)
```
